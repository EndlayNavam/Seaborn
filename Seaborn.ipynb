```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Seaborn has some built in datasets

```python
#total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#setting themes for plots
sns.set_theme()
```

```python
# visualize the data
#smoker is blue
sns.relplot(x='total_bill', y='tip', data=tips, col= 'time',hue = 'smoker', style = 'smoker',kind='scatter')
```

```python
# load iris datasets
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

Scatter plot

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris)
```

```python
sns.scatterplot(x='sepal_length', y='petal_width', hue='species', data=iris)
```
